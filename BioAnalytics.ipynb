```python
def checkPalindrome(DNA):
  converter = DNA.maketrans("ATGC", "TACG") # find the complementary strand
  complementary_strand = DNA.translate(converter)
# reverses the complementary strand
  palindrome_check = complementary_strand[::-1]
  if DNA == palindrome_check:
    return print(f"The given DNA sequence {DNA} is true")
  else:
    return print ((f"The given DNA sequence {DNA} is false"))
```

Bacterial Growth Equation

```python
N=5 #Number of bacterial Population
E=np.zeros(N) #creating an array of shape 5 using zeros function
print(E)
```

```python
E[0] = 10000.
E[1] = 2*E[0]
E[2] = 2*E[1]
E[3] = 2*E[2]
E[4] = 2*E[3]
print("Number of bacteria: ", E)
```

```python
N = 5
E = np.zeros(N)
E[0] = 10000.
for i in range(1,N):
  E[i]=2*E[i-1]
print("Number of bacteria: ", E)
```

```python
from pylab import *
plot(E, 'o-')
xlabel('Generation')
ylabel('Population size, E')
title("Modeled bacterial growth")
show()
```

```python
N=15
E=np.zeros(15)
t=np.zeros(15)
dt=20 #min
E[0] = 10000 #starting number of bacteria
for i in range(1,N):
  E[i]=2*E[i-1]
  t[i] = t[i-1] + dt
print(E) #the number of bacterial cells after 15 generations
print(t) #the time after 15 generations
plot(E, 'o-')
xlabel('Time(minutes)')
ylabel('Population size, E')
title("Modeled bacterial growth")
show()
```

```python
from pylab import *
N = 40
dt = 20 # generation time minutes
K = 8e9
E = np.zeros(N)
t = np.zeros(N)
E[0] = 10000. #initial condition
for n in range(1, N):
  E[n] = E[n-1] + (1 - E[n-1]/K)*E[n-1]
  t[n] = t[n-1] + dt
```

```python
plot(t, E, 'o-')
xlabel('Time(minutes)')
ylabel('Population size')
title("Modeled bacterial growth")
show()
```

```python
def plot_box_swarm(data, y_axis_label,x_labels, plot_title, figure_name):
  """Plot box-plot and swarm plot for data list.
      Args:
data (list of list): List of lists with data to be plotted.
y_axis_label (str): Y- axis label.
x_labels (list of str): List with labels of x-axis.
plot_title (str): Plot title.
figure_name (str): Path to output figure.
"""
  sns.set(color_codes=True)
  sns.set_style("white")
  plt.figure(figsize=(20, 10))
# add title to plot
  plt.title(plot_title)
  my_colors =['salmon','salmon','salmon','salmon','salmon','mediumturquoise','mediumturquoise','mediumturquoise','mediumturquoise','mediumturquoise']
  ax = sns.violinplot(data=data_1,width=0.8,showfliers=False,palette=my_colors)
# y-axis label
  ax.set(ylabel=y_axis_label)
  # write labels with number of elements
#ax.set_xticklabels(["{} (n={})".format(l, len(data[x])) for x, l inenumerate(x_labels)], rotation=10)
# write figure file with quality 400 dpi
  plt.savefig(figure_name, bbox_inches='tight', dpi=1200)
  plt.show()
```
